```python
import pandas as pd
import numpy as np
```

```python
def renameColumns(df, fname = None):
    data_path='input/BDB2022-plots/'
    df.index.name = 'playerName'
    names_dict = {'count': 'totalSnaps',
        'mean': 'avgFireTime',
        'median':'medFireTime'
    }

    to_rename = {key: names_dict[key] for key in names_dict.keys() if key in df.columns}
    df = df.round(2).rename(columns = to_rename)
    if fname is not None:
        # df.to_csv(data_path + fname)
        pass

    return df
    
```

```python
gunners_array = np.array(
    [
        [38707, 45648, 42718, 43988, 45021, 33234, 38596, 39222,],
        [45020, 46221, 39975, 46671, 43495, 42094, 47915, 45797,],
        [46233, 47822, 45657, 41354, 43488, 46522, 48241, 46263],
        [42397, 46992, 46729, 42359, 46178, 43396, 46724, 46349],
    ]
)
```

```python
df = pd.read_csv('firstFire2019_2.csv')
dfplays = pd.read_csv('input/nfl-big-data-bowl-2022/plays.csv')
games = pd.read_csv('input/nfl-big-data-bowl-2022/games.csv')
players= pd.read_csv('input/nfl-big-data-bowl-2022/players.csv')

dfplays = dfplays[['gameId','playId','quarter','down', 'gameClock','playResult', 'playDescription']]
```

```python
df = df.merge(players[['nflId','Position','displayName']])
df = df[df.nflId.isin(gunners_array.ravel())]
```

```python
df1 = df[df['vises'] == 1]
df2 = df[df['vises'] == 2]
```

```python
df1[df1.displayName.str.contains('Facyson')].sort_values(by = 'fireTimeS')
```

```python
pdf = df1.groupby('displayName')['fireTimeS'].agg(['mean', 'median', 'count']).sort_values(by = 'mean')
pdf.pipe(renameColumns,fname = 'gunners1.csv')
```

```python
pdf = df2.groupby('displayName')['fireTimeS'].agg(['mean', 'median', 'count']).sort_values(by = 'median')
pdf[pdf['count']>=7].pipe(renameColumns, fname = 'gunners2.csv')
```
